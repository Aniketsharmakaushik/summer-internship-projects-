```python
import pandas as pd
import numpy as np
df= pd.read_csv("salary_dataset.csv")





# Display first 5 rows
print("First 5 rows:")
print(df.head())

# Display dataset information
print("\nDataset Info:")
print(df.info())

# Summary statistics
print("\nSummary Statistics:")
print(df.describe())

# Mean of numeric columns
print("\nMean:")
print(df.mean(numeric_only=True))

# Median of numeric columns
print("\nMedian:")
print(df.median(numeric_only=True))

# Standard deviation
print("\nStandard Deviation:")
print(df.std(numeric_only=True))

# Variance
print("\nVariance:")
print(df.var(numeric_only=True))

# Minimum values
print("\nMinimum:")
print(df.min(numeric_only=True))

# Maximum values
print("\nMaximum:")
print(df.max(numeric_only=True))



# Missing values
print("\nMissing Values:")
print(df.isnull().sum())

# NumPy analysis
numeric_data = df.select_dtypes(include=np.number)

print("\nNumPy Statistics:")
print("Mean:", np.mean(numeric_data, axis=0))
print("Median:", np.median(numeric_data, axis=0))
print("Standard Deviation:", np.std(numeric_data, axis=0))
print("Minimum:", np.min(numeric_data, axis=0))
print("Maximum:", np.max(numeric_data, axis=0))
```
